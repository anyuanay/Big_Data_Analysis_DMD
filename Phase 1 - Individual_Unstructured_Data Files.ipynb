<b> <u> Individual File Data Extraction </u> </b>

This is the first method, which I have used for Data Extraction. In this method, I have first extracted each drug discovery clinical study data of the Top 10 clinical studies with DMD diseease into a separate file and then have used the Python Dom Minidom Parser to extract required fields. Each cell in this notebook represents an individual clinical study. I have extracted all fields in the clinical study, so that I can get a eloborate look over the elemnts of the clinical study, which would help me in designing the Drug Discovery Study Ontology. 

```python
# The command in this cell calls the DOM minidom Parser
from xml.dom import minidom
```

```python

# Parsing the xml file using minidom Parser and assiging this parsing a variable called doc, which will be used for
# calling the variable in the latter parts of the code.
doc = minidom.parse("/Users/Tarun/Documents/Courses/Fall_2016/Research/Data/search_result/NCT00592553.xml")


# Parsing the XMl doc to get Rank information about the study
clinicalstudy = doc.getElementsByTagName("clinical_study")
for clinical_study in clinicalstudy:
    ranking = clinical_study.getAttribute("rank")
    print("Rank: %s"% (ranking))
print ("**********************************")
 


# Parsing the XMl doc to get NCT ID Number of the study
NCTid = doc.getElementsByTagName("nct_id")[0]
print("NCT Id: %s" % (NCTid.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get the Title of the study
title = doc.getElementsByTagName("official_title")[0]
print("Title: %s" % (title.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get the Sponsors Information of the study
sponsor = doc.getElementsByTagName("sponsors")
for sponsors in sponsor:
    LeadSponsor = sponsors.getElementsByTagName("agency")[0]
    SponsorClassType = sponsors.getElementsByTagName("agency_class")[0]
    print("Lead Sponsor: %s, Sponsor Class Type: %s" % (LeadSponsor.firstChild.data, SponsorClassType.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get source of the study
source = doc.getElementsByTagName("source")[0]
print("Source: %s" % (source.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get Start Date of the Clinical Trial study
startDate = doc.getElementsByTagName("start_date")[0]
print("Start Date: %s" % (startDate.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get Completion Date of the Clinical Trial study
completionDate = doc.getElementsByTagName("completion_date")[0]
print("Completion Date: %s" % (completionDate.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get Overall Status of the Clinical Study
status = doc.getElementsByTagName("overall_status")[0]
print("Status: %s" % (status.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get phase at which the study is in.
phase = doc.getElementsByTagName("phase")[0]
print("Phase: %s" % (phase.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get information about the number of patients enrolled in the study
patientsEnrolled = doc.getElementsByTagName("enrollment")[0]
print("Patients Enrolled: %s" % (patientsEnrolled.firstChild.data))   
print ("**********************************")


# Parsing the XMl doc to get Study Type information about the Clinical Trial study
studyType = doc.getElementsByTagName("study_type")[0]
print("Study Type: %s" % (studyType.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to get Study Design information about the Clinical Trial study
studyDesign = doc.getElementsByTagName("study_design")[0]
print("Study Design: %s" % (studyDesign.firstChild.data))
print ("**********************************")


# Parsing the XMl doc to retrieve the primary outcome of the study
print ("-------- Primary Outcome ---------")
primaryOutcome = doc.getElementsByTagName("primary_outcome")
for primary_outcome in primaryOutcome:
    measure = primary_outcome.getElementsByTagName("measure") [0]
    timeFrame = primary_outcome.getElementsByTagName("time_frame") [0]
    safetyIssue = primary_outcome.getElementsByTagName("safety_issue") [0]
    print("Primary Outcome Measure:  %s" % (measure.firstChild.data))
    print("Time Frame: %s" % (timeFrame.firstChild.data))
    print("Safety Issue: %s" % (safetyIssue.firstChild.data))
print ("**********************************")    


# Parsing the XMl doc to retreive secondary outcomes of the study. 
secondaryOutcomeStudy = doc.getElementsByTagName("secondary_outcome")
for secondary_outcome in secondaryOutcomeStudy:
#Each secondary outcome is seperated a tag line.
    print ("------ Secondary Outcome ------")    
    Measure = secondary_outcome.getElementsByTagName("measure")[0]
    TimeFrame = secondary_outcome.getElementsByTagName("time_frame")[0]
    SafetyIssue = secondary_outcome.getElementsByTagName("safety_issue")[0]
    print("Measure: %s" % Measure.childNodes[0].data)
    print("Time Frame: %s" % TimeFrame.childNodes[0].data)
    print("Safety Issue: %s" % SafetyIssue.childNodes[0].data)
print ("**********************************")    


# Parsing the XMl doc to retreive the detailed description of the study.
studyDescription = doc.getElementsByTagName("detailed_description")
for detailed_description in studyDescription:
    description = detailed_description.getElementsByTagName("textblock") [0]
    print("What is the Study about: %s" % (description.firstChild.data))
print ("**********************************")
    
# Parsing the XMl doc to retreive Arm group of the study.     
armGroup = doc.getElementsByTagName("arm_group")
for arm_group in armGroup:
#Each arm group is seperated a tag line.
    print ("-------- Arm Group ---------")
    armGroupLabel = arm_group.getElementsByTagName("arm_group_label") [0]
    armGroupType = arm_group.getElementsByTagName("arm_group_type") [0]
    print("Arm Group Label:  %s" % (armGroupLabel.firstChild.data))
    print("Arm Group Type: %s" % (armGroupType.firstChild.data))
print ("**********************************")  

# Parsing the XMl doc to retreive Intervention of the study, which describes the drug name of the study.     
Intervention = doc.getElementsByTagName("intervention")
for intervention in Intervention:
#Each intervention is seperated a tag line.
    print ("-----------Interventions---------")
    interventionType = intervention.getElementsByTagName("intervention_type") [0]
    interventionName = intervention.getElementsByTagName("intervention_name") [0]
    interventionDesc = intervention.getElementsByTagName("description") [0]
    interventionArmGroupLabel = intervention.getElementsByTagName("arm_group_label") [0]
    print("Intervention Type:  %s" % (interventionType.firstChild.data))
    print("Intervention Name:  %s" % (interventionName.firstChild.data))
    print("Description:  %s" % (interventionDesc.firstChild.data))
    print("Arm Group Type: %s" % (interventionArmGroupLabel.firstChild.data))
print ("**********************************") 
```
