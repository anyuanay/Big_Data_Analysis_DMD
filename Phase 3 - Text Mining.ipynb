<b><u> Text Mining </u></b>

In this Jupyter Notebook, I have focused on using text mining to parse the unstructured data file and break down long complex sentenses into smaller chunks of valuable information. In this notebook, I have used two kinds of algorithms to text mine the data. Those algorithms are:

Algorithm 1: Text Mining to extract Nouns.

Algorithm 2: Text Mining to extract Noun Phrases.

For algorithm 1, I have used lambda function along with Parts of Speech tagging to extact only nouns.

For algorithm 2, I have used an in-built NLTK function called Textblob for extracting Noun Phrases.

In this notebook too, I have planned to use text mining on regular expression parsing, element tree parsing, and Dom Minidom parsing.

The different types of text mining examples I have used are shown below:

```python
# This cell uses algorithm 1 to extract nouns using element tree parser.

import nltk

from xml.dom import minidom
import nltk

from xml.dom.minidom import parse, Node
from textblob import TextBlob

# In the below commands I have used element tree parser using nodes.

xmltree=parse("/Users/Tarun/Documents/Courses/Fall_2016/Research/Data/search_result/NCT00592553.xml")
for node1 in xmltree.getElementsByTagName('official_title'):
    for node2 in node1.childNodes:
        if node2.nodeType == Node.TEXT_NODE:
            print(node2.data)

# The code below is the algorithm 1, that I have designed fro extracting only nouns.
is_noun = lambda pos: pos[:2] == 'NN'
tokenized = nltk.word_tokenize(node2.data)
nouns = [word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 
print (nouns)

```

```python
# This cell uses algorithm 2 to extract Noun Phrases using the Element Tree Parse as above.

from xml.dom import minidom

import nltk


from xml.dom.minidom import parse, Node

# As I am using algorithm 2, I need to call Textblob, which is an in-built function for text mining in python.

from textblob import TextBlob

# In the below commands I have used element tree parser using nodes.
xmltree=parse("/Users/Tarun/Documents/Courses/Fall_2016/Research/Data/search_result/NCT00592553.xml")
for node1 in xmltree.getElementsByTagName('official_title'):
    for node2 in node1.childNodes:
        if node2.nodeType == Node.TEXT_NODE:
            print(node2.data)

# In the below commands I have used element tree parser using nodes.
for node3 in xmltree.getElementsByTagName('start_date'):
    for node4 in node3.childNodes:
        if node4.nodeType == Node.TEXT_NODE:
            print(node4.data)


# For algorithm 2, I have used textblob as below to call the node data that I would like to turn 
# the sentences as Noun Phrases.
nlpblob = TextBlob(node2.data)
        
# The noun phrase function would can be called along with textblob.
nlpblob.noun_phrases

```

```python
# In this cell I have used algorithm 2 (Textblob) for extracting data using Regular Expressions

import re
from textblob import TextBlob

# The commands below extracts the title and start date using regular expressions.
text=open("/Users/Tarun/Documents/Courses/Fall_2016/Research/Data/search_result/NCT00592553.xml").read()
found=re.findall("<official_title>(.*)</official_title>",text)
for title in found:
    print (title)

f2 = re.findall("<start_date>(.*)</start_date>", text)
for startDate in f2:
    print (startDate)

    
# I have used textblob to extract noun phrases from the title using the commands below.
nlpblob = TextBlob(title)
nlpblob.noun_phrases    
```
